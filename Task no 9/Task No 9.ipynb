```python
import pandas as pd

dataset = pd.read_csv('road_accident_dataset.csv')

print("=" * 80)
print("1. TOP 5 ROWS OF DATASET:")
print("=" * 80)
print(dataset.head())
print("\n" + "=" * 80)

print("\n2. BOTTOM 5 ROWS OF DATASET:")
print("=" * 80)
print(dataset.tail())
print("\n" + "=" * 80)

print("\n3. DATASET DIMENSIONS:")
print("=" * 80)
print(f'Rows: {dataset.shape[0]}')
print(f'Columns: {dataset.shape[1]}')
print("=" * 80)

print("\n4. NULL VALUES IN EACH COLUMN:")
print("=" * 80)
null_counts = dataset.isnull().sum()
print(null_counts)
print(f'\nTotal null values in dataset: {null_counts.sum()}')
print("=" * 80)

print("\n5. DATA TYPES OF EACH COLUMN:")
print("=" * 80)
print(dataset.dtypes)
print("=" * 80)

print("\n6. ADDITIONAL EXPLORATORY ANALYSIS:")
print("=" * 80)

print("\nA. BASIC STATISTICS FOR NUMERICAL COLUMNS:")
print("-" * 40)
print(dataset.describe())

print("\nB. DATASET INFORMATION:")
print("-" * 40)
dataset.info()

categorical_columns = ['Country', 'Month', 'Day of Week', 'Urban/Rural', 'Road Type', 
                       'Weather Conditions', 'Driver Age Group', 'Driver Gender', 
                       'Accident Severity', 'Road Condition', 'Accident Cause', 'Region']
print("\nC. UNIQUE VALUES IN KEY CATEGORICAL COLUMNS:")
print("-" * 40)
for col in categorical_columns:
    print(f"{col}: {dataset[col].nunique()} unique values")
    print(dataset[col].value_counts().head())
    print()
```
