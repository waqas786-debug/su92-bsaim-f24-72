```python
import pandas as pd
import numpy as np

data = pd.read_csv('road_accident_dataset.csv')

print('We have {} rows.'.format(data.shape[0]))
print('We have {} columns'.format(data.shape[1]))

print(np.sum(pd.isnull(data)))
print("\nTotal null values:", np.sum(pd.isnull(data).sum()))

categorical_columns = ['Country', 'Month', 'Day of Week', 'Time of Day', 'Urban/Rural', 
                       'Road Type', 'Weather Conditions', 'Visibility Level', 
                       'Driver Age Group', 'Driver Gender', 'Vehicle Condition',
                       'Accident Severity', 'Road Condition', 'Accident Cause', 'Region']
for col in categorical_columns[:5]:
    print(f"\n{col}: {data[col].unique()}")

x = data.iloc[:, 0:-1]
y = data.iloc[:, -1]

cat_columns = x.select_dtypes(['object']).columns
print(f"\nCategorical columns to convert: {list(cat_columns)}")

for col in cat_columns:
    if col in x.columns:
        x[col] = pd.factorize(x[col])[0]

print("\nData types after conversion:")
print(x.dtypes.value_counts())

print("\nFirst few rows of transformed X:")
print(x.head())
```
