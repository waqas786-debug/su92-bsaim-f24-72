```python
from flask import Flask, render_template_string
import requests

app = Flask(__name__)

@app.route('/')
def home():
    url = "https://official-joke-api.appspot.com/random_joke"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        html_template = f'''
        <div style="text-align:center; margin-top:50px; font-family:Arial;">
            <h1>Lab 7: Random Joke App</h1>
            <div style="border:1px solid #ccc; padding:20px; display:inline-block; border-radius:10px;">
                <h3>Setup: {data['setup']}</h3>
                <p style="color:blue; font-weight:bold;">Punchline: {data['punchline']}</p>
                <button onclick="window.location.reload();">Get Another Joke</button>
            </div>
        </div>
        '''
        return render_template_string(html_template)
    return "Error: Could not fetch joke"

if __name__ == '__main__':
    app.run(debug=True, port=5001)
```
