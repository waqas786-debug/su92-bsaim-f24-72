```python
import cv2
import os

if not os.path.exists("sample.jpg"):
    import numpy as np
    dummy_img = np.zeros((500, 500, 3), dtype=np.uint8)
    cv2.putText(dummy_img, "Sample Image", (100, 250), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)
    cv2.imwrite("sample.jpg", dummy_img)

image = cv2.imread("sample.jpg")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
resized_image = cv2.resize(image, (300, 300))

cv2.rectangle(resized_image, (50, 50), (200, 200), (0, 255, 0), 2)
cv2.circle(resized_image, (150, 150), 40, (0, 0, 255), -1)

_, thresh = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
edges = cv2.Canny(gray_image, 100, 200)

contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(resized_image, contours, -1, (255, 0, 0), 2)

cv2.imshow("Original", image)
cv2.imshow("Processed", resized_image)
cv2.imshow("Edges", edges)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
