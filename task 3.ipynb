```python
def dfs_water_jug(capacity_x, capacity_y, target):
    visited = set()
    stack = [((0, 0), [])]

    while stack:
        (x, y), path = stack.pop()

        if x == target or y == target:
            for step in path + [(x, y)]:
                print(step)
            return True

        if (x, y) in visited:
            continue
        visited.add((x, y))

        moves = [
            ((capacity_x, y), "Fill Jug 1"),
            ((x, capacity_y), "Fill Jug 2"),
            ((0, y), "Empty Jug 1"),
            ((x, 0), "Empty Jug 2"),
            ((x - min(x, capacity_y - y), y + min(x, capacity_y - y)), "Pour 1 to 2"),
            ((x + min(y, capacity_x - x), y - min(y, capacity_x - x)), "Pour 2 to 1")
        ]

        for (nx, ny), rule in moves:
            if (nx, ny) not in visited:
                stack.append(((nx, ny), path + [((x, y), rule)]))
    return False

dfs_water_jug(4, 3, 2)
```
